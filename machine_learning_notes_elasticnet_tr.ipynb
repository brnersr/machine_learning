## MAKİNE ÖĞRENMESİNE GİRİŞ
### DOĞRUSAL REGRESYON MODELİ- ElasticNet

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv("/Users/bernaersari/Downloads/Hitters.csv")
```

```python
df # beyzbol oyuncularının çeşitli özeLliklerine göre ücretini belirleyen bir model kurmak istiyoruz.
```

```python
df=df.dropna() #boş değerleri silelim.
```

```python
df.select_dtypes(include='object').columns
```

```python
dms=pd.get_dummies(df[['League', 'Division', 'NewLeague']]) # one-hot encoding ile kategorik değişkenleri 0-1 değişkenlere dönüştürelim.
```

```python
y=df['Salary'] #bulmak istediğimiz hedef değişkenimiz olan 'Salary' bağımlı değişkenimizi y olarak tanımlayalım.
```

```python
X_=df.drop(['Salary','League', 'Division', 'NewLeague'], axis=1) 
#dataframeden encoding yapılan değişkenleri ve bağımlı değişkeni silip dataframedeki verileri X_ olarak tanımlayalım.
```

```python
X_ #bağımsız değişkenlere göz atalım.
```

```python
X=pd.concat([X_,dms[['League_N', 'Division_W','NewLeague_N']]],axis=1) 
#son olarak tüm bağımsız değişkenleri bir araya getirelim. 
#Dummy değişkenler aynı bilgileri içerdiğinden her birinden bir tane değişkeni bağımsız değişken grubuna ekliyoruz.
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                               test_size=0.25,
                                               random_state=42) # veri setinin yüzde 25 test seti olarak ayrıldı.
```

```python
ElasticNet_model=ElasticNet().fit(X_train,y_train)
#ElasticNet modeli oluşturup,modeli eğitelim. 
```

```python
ElasticNet_model
```

```python
ElasticNet_model.coef_ #modeli eğittiğimizde değişkenlere uygulanan katsayılar
```

```python
pd.Series(ElasticNet_model.coef_,index=X_train.columns).sort_values(ascending=False)
#hangi değişkenlere hangi katsayılar uygulanmış bir bakalım. 
#Burada hangi değişkenlerin beyzbol oyuncusunun maaşına etki ettiği görülüyor. 
#Örneğin, hits değişkeninin maaşa olumlu katkısı olurken;CHits değişkeninin maaşa etkisi negatif olduğu görülüyor.
```

```python
ElasticNet_model.intercept_ #sabit katsayı
```

```python
y_pred=ElasticNet_model.predict(X_test) # test için ayırdığımız veri setini modelimiz tahmin yapsın.
```

```python
hata1=np.sqrt(mean_squared_error(y_test,y_pred)) #Optimizasyon yapılmadan tahmin edilen değerlerimizin hatasına bakalım. 
```

```python
hata1
```

#### MODEL TUNING


```python
lambda_numbers=np.sqrt(10**np.linspace(10,-1,50))
lambda_numbers
```

OPTİMUM LAMBDA DEĞERİNİ BULMAK İÇİN ElasticNetCV KULLANALIM.

```python
ElasticNet_model_cv=ElasticNetCV(alphas=lambda_numbers,cv=10).fit(X_train,y_train)
```

```python
ElasticNet_model_cv.alpha_ # alpha seçimi için bir sayı kümesi vermiştik. 
#Bu sayı kümesi içindne en uygun olanını modelimiz seçti
```

```python
#tune edilmiş final modelimizi kuralım.
```

```python
ElasticNet_model_final=ElasticNet(alpha=ElasticNet_model_cv.alpha_).fit(X_train,y_train)
```

```python
y_pred=ElasticNet_model_final.predict(X_test) #modelimiz test veri seti üzerinden tahminleme yapsın.
```

```python
hata2=np.sqrt(mean_squared_error(y_test,y_pred)) #yapılan tahmnin hatasına bakalım.
```

```python
hata2
```

```python
print(hata1,hata2) # tune ettikten sonra hatamız arttı. bu bazen görülen bir durum.
```

```python
# alpha seçimi bir sayı kümeden modeli eğitelim.
```

```python
ElasticNet_model_cv2=ElasticNetCV(cv=10).fit(X_train,y_train)
```

```python
ElasticNet_model_cv2.alpha_ #modelimiz kendisi en iyi alpha değerini seçti.
```

```python
ElasticNet_model_final2=ElasticNet(alpha=ElasticNet_model_cv2.alpha_).fit(X_train,y_train)
#alpha değeri için kendisine bir veri seti kümesi vermeden seçilmiş en iyi alpha değeri ile modeli eğitelim.
```

```python
y_pred=ElasticNet_model_final2.predict(X_test)
```

```python
hata3=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
hata3
```

```python
print(hata1,hata2,hata3)
```

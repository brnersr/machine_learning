### DOĞRUSAL OLMAYAN REGRESYON MODELİ- KNN (K-NEAREST NEIGHBORS MODEL)

```python
import warnings 
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/bernaersari/Downloads/Hitters.csv")
```

```python
df # beyzbol oyuncularının çeşitli özeLliklerine göre ücretini belirleyen bir model kurmak istiyoruz.
```

```python
df=df.dropna() #boş değerleri silelim.
```

```python
df.select_dtypes(include='object').columns
```

```python
dms=pd.get_dummies(df[['League', 'Division', 'NewLeague']]) # one-hot encoding ile kategorik değişkenleri 0-1 değişkenlere dönüştürelim.
```

```python
y=df['Salary'] #bulmak istediğimiz hedef değişkenimiz olan 'Salary' bağımlı değişkenimizi y olarak tanımlayalım.
```

```python
X_=df.drop(['Salary','League', 'Division', 'NewLeague'], axis=1) 
#dataframeden encoding yapılan değişkenleri ve bağımlı değişkeni silip dataframedeki verileri X_ olarak tanımlayalım.
```

```python
X_ #bağımsız değişkenlere göz atalım.
```

```python
X=pd.concat([X_,dms[['League_N', 'Division_W','NewLeague_N']]],axis=1) 
#son olarak tüm bağımsız değişkenleri bir araya getirelim. 
#Dummy değişkenler aynı bilgileri içerdiğinden her birinden bir tane değişkeni bağımsız değişken grubuna ekliyoruz.
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                               test_size=0.25,
                                               random_state=42) # veri setinin yüzde 25 test seti olarak ayrıldı.
```

__MODEL & TAHMİN__

```python
knn_model=KNeighborsRegressor().fit(X_train,y_train) #en yakın komşu modeli kuralım.
```

```python
y_pred=knn_model.predict(X_test) #modelimiz tahminlerini yapar.
```

```python
test_hatasi=np.sqrt(mean_squared_error(y_test,y_pred)) #test_hatası değişkenine hata değerini atayalım.
```

__MODEL_TUNING__

```python
knn_params={"n_neighbors":np.arange(1,30,1),
            'p':[1,3],
           "leaf_size":[10,20,30]}  #hiperparametre seçimleri için birkaç veri seti verelim.
```

```python
knn_model_tuned=GridSearchCV(knn_model,knn_params,cv=10).fit(X_train,y_train) #gridsearch ile verdiğimiz hiperparametre veri setlerine göre modeli eğitelim.
```

```python
knn_model_tuned.best_params_ #gridsearchcv ile seçilen rn iyi hiperparametrelere bakalım.
```

```python
knn_model_final=KNeighborsRegressor(n_neighbors=knn_model_tuned.best_params_["n_neighbors"],
                                   p=knn_model_tuned.best_params_["p"],
                                   leaf_size=knn_model_tuned.best_params_["leaf_size"]
                                   ).fit(X_train,y_train)
#Seçilen en iyi hiperparametreler ile modelimizi eğitelim(fit edelim)
```

```python
y_pred=knn_model_final.predict(X_test) #modelimiz tahminlesin.
```

```python
np.sqrt(mean_squared_error(y_test,y_pred)) #modelimizin hatasına bakalım.
```
